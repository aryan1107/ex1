```python
import pandas as pd
import matplotlib
import numpy as np
import pandas as pd
pd.options.mode.chained_assignment = None
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter(action="ignore", category=FutureWarning)
```

```python
# reading csv
df = pd.read_csv('/Users/aryan/Desktop/shopify/2019 Winter Data Science Intern Challenge Data Set - Sheet1.csv')
df.head()
```

```python
average_order_value_Sum = df['order_amount'].sum()
item_Sum = df['total_items'].sum()

print(average_order_value_Sum)
print(item_Sum)
```

```python
# Calculating the mean (AOV) by dividing total order amount by the total items amount.
AOV = average_order_value_Sum/item_Sum

#Rounding AOV to two decimal places.
round(AOV, 2)
```

```python
# Now doing exploritory analysis
```

```python
#plotting on canvas to find out how data is represented
fig, ax = plt.subplots(figsize=(9,6))


ax.scatter(df['total_items'], df['order_amount'])

ax.set_title('Total Item vs Order amount')
ax.set_xlabel('Total Item')
ax.set_ylabel('Order amount')
```

```python
# Removing the outlier
df_3 = df[(df['order_amount'] < 20000)]
#plotting on canvas to find out what's wrong
fig, ax = plt.subplots(figsize=(9,6))


ax.scatter(df_3['total_items'], df_3['order_amount'])

ax.set_title('Total Item vs Order amount')
ax.set_xlabel('Total Item')
ax.set_ylabel('Order amount')
```

```python
# Box plot
plt.figure(figsize=(9,12))
sns.boxplot('total_items', 'order_amount', data=df_3)
```
